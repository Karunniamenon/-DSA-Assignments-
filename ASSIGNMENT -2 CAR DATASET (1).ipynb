## 1.Read the dataset to the python environment.

```python
cars_data=pd.read_csv("cars_data.csv")
```

## 2.Check for the null values present in the dataset.

```python
cars_data.info()## SHOWS, IT HAS 1000 ENTRIES AND ALL HAS VALUES (SINCE VALUE OF NON NULL==1000)
```

```python
### METHOD-2 USING isnull.sum
```

```python
cars_data.isnull().sum() ## HERE SUM WE GOT AFTER THE OPERATION IN EACH PARAMETER IS ZERO, HENCE IT DO NOT CONTAIN ANY NULL VALUES.
```

```python
### METHOD-3 USING isna
```

```python
cars_data.isna().values ##  HERE WE CANNOT FIND ANY TRUE VALUE IN THE DATASET . BUT IT IS NOT CLEAR SINCE IT IS A LARGER SET
```

```python
cars_data.isna().sum() ##  HERE WE TAKE THE SUM OF NULL VALUES OF ALL PARAMETERS , ALL VALUES COUNTED ARE = ZERO,
                       ##  HENCE WE CAN CONCLUDE THAT THE DATASET DOESNOT CONTAIN ANYNULL VALUES.
```

```python
### METHOD-4 USING isnull
```

```python
cars_data.isnull().values.any() ##  WE GET NULL VALUE OF ALL PARAMETERS FALSE, REFLECT THERE IS NO NULL VALUES IN DATASET
```

## 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
cars_data['Buyer Gender'].value_counts()
```

### METHOD-1 USING COUNT PLOT

```python
plt.figure(figsize=(9,6))
 
sns.countplot(cars_data['Buyer Gender'], palette='Set2')
```

```python
plt.figure(figsize=(9,6))
sns.set_theme(style="darkgrid")
sns.countplot(cars_data['Buyer Gender'], facecolor=(0,0,0,0), linewidth=5, edgecolor=sns.color_palette("dark",3))
```

### METHOD 2 USING plt

```python
figsize=(8,6)
cars_data['Buyer Gender'].value_counts().plot(kind='bar',rot=30,color=['magenta','lightgreen'])
plt.xlabel("BUYER GENDER",labelpad=14)
plt.ylabel("COUNT",labelpad=12)
plt.title(" GENDER WISE SALES PARTICIPATION",fontsize=16)
```

### INFERENCE FROM THE PLOT OF SALES PARTICIPATION:-
     With respect to gender wise classification in participation of sale
            1. Women participation is above 5000, which highlights their higher participation than men.
            2. Total pariticipation is 10,000 people.

## 5. Find the least 5 cars based on their Resell price.

```python
resale_price=cars_data.sort_values(by=['Resell Price'],axis=0,ascending=[True])
```

```python
resale_price ## SORTED DATASET WITH LEAST RESELL PRICE AT FIRST.
```

```python
resale_price.head(5)
```

```python
resale=resale_price[['Make','Model','Resell Price']].head(5)
```

```python
resale## FINAL OUTPUT ACCORDING TO LEAST RESELL PRICE
```

### INFERENCES:-

      2. Least resell price didn't go below 2000.

### VISUALIZATION BETWEEN SALE PRICE AND RESELL PRICE

###  1.VISUALIZATION OF RESELL PRICE AND ITS PARAMETERS ABOVE MENTIONED

```python
plt.figure(figsize=(10,12))

plt.subplot(3,1,1)
plt.scatter(resale['Make'],resale['Resell Price'],s=500,c='blue',marker='^')
plt.title('Plot of Make vs Resell Price ',fontsize=20,c='m')
plt.xlabel('Make',fontsize=15)
plt.ylabel('Resell Price',fontsize=15)
plt.grid()

plt.subplot(3,1,2)
plt.scatter(resale['Model'],resale['Resell Price'],s=500,c='red')
plt.title('Plot of Model vs Resell Price',fontsize=20)
plt.xlabel('Model',fontsize=15)
plt.ylabel('Resell Price',fontsize=15)
plt.grid()
plt.subplot(3,1,3)
plt.scatter(resale['Make'],resale['Model'],s=500,c='blue',marker='^')
plt.title('Plot of Model VS Make ',fontsize=20,c='m')
plt.xlabel('Make',fontsize=15)
plt.ylabel('Model',fontsize=15)
plt.grid()
plt.tight_layout()
```

### 2.VISUALIZATION OF SALE PRICE AND ITS PARAMETERS ABOVE MENTIONED

```python
plt.figure(figsize=(10,12))

plt.subplot(3,1,1)
sns.swarmplot(x='Make',y='Sale Price',s=25,hue='Make',data=sale)
plt.title('Plot of Make vs Sale Price',fontsize=20)
plt.xlabel('Make',fontsize=15)
plt.ylabel('Sale Price',fontsize=15)
plt.grid()
plt.subplot(3,1,2)
sns.swarmplot(x='Model',y='Sale Price',s=25,hue='Model',data=sale)
plt.title('Plot of Model vs Sale Price',fontsize=20)
plt.xlabel('Model',fontsize=15)
plt.ylabel('Sale Price',fontsize=15)
plt.grid()

plt.subplot(3,1,3)
sns.swarmplot(x='Make',y='Model',s=25,hue='Model',data=sale)
plt.title('Plot of Make vs Model',fontsize=20)
plt.xlabel('Make',fontsize=15)
plt.ylabel('Model',fontsize=15)
plt.grid()

plt.tight_layout()
```
